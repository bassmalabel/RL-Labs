This lab focuses on solving the system (Bellman optimality equation) using the fixed-point iteration method.
- To solve a system using the fixed-point iteration method:
1. Initialization: Start by initializing ùëâ1 and ùëâ2 with arbitrary values (e.g., 0).
2. Iterative Update: Iteratively update ùëâ1 and ùëâ2 using the Bellman optimality equations

```python
import numpy as np
v_s1=0
v_s2=0
v_s1_pre=-1
v_s2_pre=-1
i=0
while(v_s1!=v_s1_pre or v_s2!=v_s2_pre) :
    print(f'Iteraton {i} v_s1={v_s1}, v_s2={v_s2}')
    v_s1_pre=v_s1
    v_s2_pre=v_s2
    v_s1= np.max([-0.4+0.56*v_s1_pre+0.24*v_s2_pre, -0.1+0.64*v_s1_pre+0.16*v_s2_pre] )
    v_s2= np.max([-0.475+0.72*v_s1_pre+0.08*v_s2_pre , 0.4*v_s1_pre+0.4*v_s2_pre ])
    i+=1
```

```python
print(f'Iteraton {i} v_s1={v_s1}, v_s2={v_s2}')
```

After calculating the optimal value function, we can derive the optimal policy: 
- we compute the value of each action in each state and then we compare. 
- if we choose the action with maximum value in each state we will fall on the optimal policy. 

```python
v_s1_a1= -0.4+0.56*v_s1+0.24*v_s2 
v_s1_a2=-0.1+0.64*v_s1+0.16*v_s2
v_s2_a1=-0.475+0.72*v_s1+0.08*v_s2
v_s2_a2=0.4*v_s1+0.4*v_s2 

print(f'v1_a1={v_s1_a1}, v1_a2={v_s1_a2}')
print(f'v2_a1={v_s2_a1}, v2_a2={v_s2_a2}')
```
